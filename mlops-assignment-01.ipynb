2. Verify if the id of dictionary modiy if we add new key in it?Try to find justificationbehind the result.<br>
Ans:- The dictionary is a mutable data type and whenever we perform any type of operation like insertion, updation or deletion it changes in the actual value as verified by the below code.

```python
dict = {"user1":"Yuvraj","user2":"Pankaj","user3":"Amit"}
print("Actual id:",id(dict))
dict["user4"] ="Suraj"
print("Id after insertion:",id(dict))
dict["user3"]= "Kunal"
print("Id after updation:",id(dict))
del (dict["user4"])
print("Id after deletion:",id(dict))
```

5. Write python code to print below pyramid. Please note the pyramid height must be cofigurable<br>
Ans:-

```python
length = int(input("Enter the Length :"))
def stars(length):
    for i in range(1,length + 1):
        space = ' ' * (length - i)
        star = '*' * i
        print(space + star)

stars(length)
```
